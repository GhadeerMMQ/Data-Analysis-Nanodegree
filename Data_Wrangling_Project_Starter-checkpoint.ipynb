```python
neetflix_titles_raw=pd.read_csv("D:\corpo\DFOnetflix_titles.csv")
neetflix_titles_raw.info()
display(neetflix_titles_raw.tail(20))
```

#### Dataset 2

Type: API

Method: The data was gathered using the "API" method from The movie database TMDb.

Dataset variables:
'title', 'popularity', 'release_year', 'vote_average', 'type', 'primary_genre'

```python
titles = neetflix_titles_raw['title'].head(600)
results = []
print("Starting API data collection...")

for title in titles:
    try:
        url = f"https://api.themoviedb.org/3/search/movie?api_key={api_key}&query={title}"
        response = requests.get(url)
        
        # Check if request was successful
        if response.status_code == 200:
            data = response.json()
            if data['results']:
                # Extract only the 4 variables
                movie = data['results'][0]
                results.append({
                    'title': title,
                    'popularity': movie.get('popularity'),
                    'release_year': movie.get('release_date', '0000')[:4],
                    'vote_average': movie.get('vote_average')
                })
        
        # RATE LIMIT: Pause for 0.25 seconds (4 requests per second)
        time.sleep(0.25)
        
    except Exception as e:
        print(f"Error fetching {title}: {e}")

# Create the Raw API dataframe
df_tmdb_raw = pd.DataFrame(results)
```

Clean the data to solve the 4 issues corresponding to data quality and tidiness found in the assessing step. **Make sure you include justifications for your cleaning decisions.**

```python
#making copies
netflix = neetflix_titles_raw.copy()
tmdb = df_tmdb_raw.copy()
```

### **Quality Issue 1: Handling Missing Values (Data Quality)**

```python
# Drop rows missing 'title' or 'type' (essential info)
netflix = netflix.dropna(subset=['title', 'type'])

# Fill missing 'country' with 'Unknown'
netflix['country'] = netflix['country'].fillna('Unknown')
```

```python
#validating
netflix.isnull().sum()
```

Justification: Keeping incomplete but relevant rows prevents losing too much data, while replacing NaN with “Unknown” maintains dataset integrity.

### **Quality Issue 2: Inconsistent Country/Rating Formats (Consistency)**

```python
#cleaning
netflix['country'] = netflix['country'].str.strip().str.title()
netflix['rating'] = netflix['rating'].str.upper().str.replace(' ', '')
```

```python
#validating
netflix['country'].value_counts().head()
```

Justification: Standardizing text ensures that the same category (e.g., “United States”) isn’t treated as multiple unique values.

### **Tidiness Issue 1: Split Multiple Genres into Separate Rows**

```python
netflix = netflix.assign(listed_in=netflix['listed_in'].str.split(', '))
netflix = netflix.explode('listed_in').reset_index(drop=True)
```

```python
#validating
print(netflix['listed_in'].head(10))
```

Justification: Each row should represent one observation (movie–genre pair). This structure allows clearer analysis by genre.

### **Tidiness Issue 2: Convert 'type' to Categorical**

```python
netflix['type'] = netflix['type'].astype('category')
```

```python
#validating
netflix.info()
```

Justification: Categorical variables improve readability and performance for analysis and modeling.

### **Remove unnecessary variables and combine datasets**


```python
netflix = netflix[['title', 'release_year', 'country', 'listed_in']]
tmdb = tmdb[['title', 'vote_average', 'popularity']]
netflix['title_clean'] = netflix['title'].str.lower().str.strip()
tmdb['title_clean'] = tmdb['title'].str.lower().str.strip()
combined = pd.merge(
    netflix,
    tmdb,
    on='title_clean',
    how='inner',
    suffixes=('_netflix', '_tmdb')
)
combined = combined.drop(columns=['title_tmdb'])
combined = combined.rename(columns={'title_netflix': 'title'})
```

```python
combined = combined.drop_duplicates(subset=['title'], keep='first')
```

## 5. Answer the research question

### **5.1:** Define and answer the research question 

*Research question:* Which movie genres on Netflix tend to have higher TMDB ratings, and how does popularity relate to these ratings?

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# For plotting style
sns.set(style="whitegrid")

# Group by genre and calculate mean rating and popularity
genre_stats = combined.groupby('listed_in').agg({
    'vote_average': 'mean',
    'popularity': 'mean'
}).sort_values('vote_average', ascending=False).reset_index()
```

```python
plt.figure(figsize=(12,6))
sns.barplot(data=genre_stats, x='vote_average', y='listed_in', palette='viridis')
plt.title("Average TMDB Rating by Genre for Netflix Movies")
plt.xlabel("Average Rating")
plt.ylabel("Genre")
plt.show()
```

*Answer to research question:* classic movies have the best quality according to TMDB users

```python
plt.figure(figsize=(12,6))
sns.scatterplot(data=combined, x='popularity', y='vote_average', hue='listed_in', alpha=0.7)
plt.title("TMDB Popularity vs Rating for Netflix Movies by Genre")
plt.xlabel("Popularity")
plt.ylabel("Rating")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

*Answer to research question:*
Movies with ratings between 6 and 8 but low popularity (0–5) indicate that moderately well-rated films on Netflix may not attract broad viewership or social attention. This suggests a disconnect between quality and popularity, highlighting the presence of lesser-known “hidden gems” in the catalog.
